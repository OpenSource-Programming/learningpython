```
score = input("Enter Score: ")
float_version = float(score)
if float_version >= 0.9:
    print("A")
elif float_version >= 0.8:
    print("B")
elif float_version >= 0.7:
    print("C")
elif float_version >= 0.6:
    print("D")
elif float_version < 0.6:
    print("F")
else:
    print("That number is out of range.  Enter a number: ")
```
