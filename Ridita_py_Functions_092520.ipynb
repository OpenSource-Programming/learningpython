```
def computepay(h,r):
    if h <= 40:
    	return h * float(r)
    elif h > 40:
    	hours_above_forty = h-40
    	return (40*float(r)) + (hours_above_forty*1.5*float(r))

hrs = input("Enter Hours:")
rate = input("Enter Rate:")
p = computepay(float(hrs), float(rate))
print("Pay",p)
```
